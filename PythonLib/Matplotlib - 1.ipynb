```python
#Bar Graph
import matplotlib.pyplot as plt
```

```python
#Bar Graph
cars=['Audi','BMW','Ford','Tesla','Jaguar','Mercedes-Benz']
data=[23,17,35,29,12,41]
plt.bar(cars,data)
plt.xlabel("Cars imported in a month")
plt.ylabel("No of cars imported in a month")
plt.title("Cars data")
plt.show()
```

```python
#Pie chart
plt.pie(data,labels=cars,autopct="%1.2f%%")
plt.show()
```

```python
#Scatter plot
boys_grades=[89,90,92,93,94]
girls_grades=[94,69,73,100,98]
grades_range=[20,40,60,80,100]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(grades_range,girls_grades,color='r')
ax.scatter(grades_range,boys_grades,color='g')
plt.show()
```

```python
#Line plot
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

```python
#Box plot
import numpy as np
```

```python
np.random.seed(10)
x1=np.random.normal(100,10,200)
x2=np.random.normal(80,30,200)
x3=np.random.normal(90,20,200)
x4=np.random.normal(70,25,200)
data1=[x1,x2,x3,x4]
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.boxplot(data1)
plt.show()
```

```python
#3D scatter plot
from mpl_toolkits import mplot3d

x=np.random.randint(100,size=(50))
y=np.random.randint(80,size=(50))
z=np.random.randint(60,size=(50))

fig=plt.figure()
ax=plt.axes(projection="3d")

ax.scatter3D(x,y,z,color="green")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('C:/Users/Aarif/Downloads/pokemon_alopez247.csv')
```

```python
#Bar graph from pokemon dataset
from scipy.stats import itemfreq
# Retrieve a list of group types against the
# number of pokemon(s) belonging to that group
type_1 = itemfreq(df.iloc[:,2])
print(type_1)
# Total number of distinct groups
type_1_grps = len(type_1)
print(type_1_grps)
# Names of group
type_1_names = type_1[:,0]
print(type_1_names)
# Pokemon count particular to each group
type_1_count = type_1[:,1]
print(type_1_count)
```

```python
type_1_grps = np.arange(type_1_grps)
bar_width = 0.5
plt.bar(type_1_grps, type_1_count, bar_width,
                 alpha = 0.5,   # tranparency factor
                 color = 'g',   # color factor
                 label='Pokemon count respective to their Type_1')
plt.legend(loc='best')
plt.xticks(type_1_grps + bar_width/2, type_1_names)
```
